```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Módulo 6 tarefa 1

#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
df = pd.read_csv('./SINASC_RO_2019.csv')
df
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python


# Configurar o estilo dos gráficos
#plt.style.use('seaborn-darkgrid')

# Criando gráficos de barras para cada variável

# 1. Gráfico de barras para APGAR1
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
df['APGAR1'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribuição de APGAR1')
plt.xlabel('Pontuação APGAR1')
plt.ylabel('Contagem')

# 2. Gráfico de barras para APGAR5
plt.subplot(2, 2, 2)
df['APGAR5'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribuição de APGAR5')
plt.xlabel('Pontuação APGAR5')
plt.ylabel('Contagem')

# 3. Gráfico de barras para SEXO
plt.subplot(2, 2, 3)
df['SEXO'].value_counts().plot(kind='bar', color=['blue', 'pink'])
plt.title('Distribuição por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Contagem')

# 4. Gráfico de barras para GESTACAO
plt.subplot(2, 2, 4)
df['GESTACAO'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribuição de Gestação')
plt.xlabel('Idade Gestacional (Semanas)')
plt.ylabel('Contagem')

# Ajustar layout e exibir os gráficos
plt.tight_layout()
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
media_apgar5_por_consultas = df.groupby('CONSULTAS')['APGAR5'].mean()

# Plotar o gráfico de barras
plt.figure(figsize=(10, 6))
media_apgar5_por_consultas.plot(kind='bar', color='skyblue', edgecolor='black')

# Configurar o título e rótulos dos eixos
plt.title('Média de APGAR5 por CONSULTAS')
plt.xlabel('CONSULTAS')
plt.ylabel('Média de APGAR5')

# Mostrar o gráfico
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x para melhor legibilidade
plt.tight_layout()
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python

# Contar frequências para cada variável
freq_sexo = df['SEXO'].value_counts()
freq_gestacao = df['GESTACAO'].value_counts().sort_index()
freq_est_civ_mae = df['ESTCIVMAE'].value_counts()
freq_consultas = df['CONSULTAS'].value_counts()

# Criar um painel com 4 gráficos
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Gráfico de barras para SEXO
freq_sexo.plot(kind='bar', ax=axes[0, 0], color=['blue', 'pink'])
axes[0, 0].set_title('Distribuição por SEXO')
axes[0, 0].set_xlabel('Sexo')
axes[0, 0].set_ylabel('Contagem')

# Gráfico de barras para GESTACAO
freq_gestacao.plot(kind='bar', ax=axes[0, 1], color='lightgreen')
axes[0, 1].set_title('Distribuição de GESTACAO')
axes[0, 1].set_xlabel('Idade Gestacional (Semanas)')
axes[0, 1].set_ylabel('Contagem')

# Gráfico de barras para ESTCIVMAE
freq_est_civ_mae.plot(kind='bar', ax=axes[1, 0], color='gold')
axes[1, 0].set_title('Estado Civil da Mãe')
axes[1, 0].set_xlabel('Estado Civil')
axes[1, 0].set_ylabel('Contagem')

# Gráfico de barras para CONSULTAS
freq_consultas.plot(kind='bar', ax=axes[1, 1], color='skyblue')
axes[1, 1].set_title('Frequência de CONSULTAS')
axes[1, 1].set_xlabel('Tipo de Consulta')
axes[1, 1].set_ylabel('Contagem')

# Ajustar layout para evitar sobreposição
plt.tight_layout()

# Exibir os gráficos
plt.show()
```
