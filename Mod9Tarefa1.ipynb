### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
# Calcular a contagem de DTNASC e as estatísticas das idades por munResNome
minimo = (
    sinasc_raw.groupby('munResNome')
    .agg({
        'DTNASC': 'count',  # Contar o número de ocorrências de DTNASC
        'IDADEMAE': ['mean', 'max', 'min'],  # Estatísticas das idades das mães
        'IDADEPAI': ['mean', 'max', 'min']   # Estatísticas das idades dos pais
            })
)
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



### Analise as respostas encontradas, tire algum insight delas, conte pra gente algo encontrado nos dados. Algo que você julgue relevante e novo pra você.

```python
import pandas as pd

# Supondo que sinasc_raw seja o seu DataFrame
# Primeiro, convertemos a coluna DTNASC para datetime
sinasc_raw['DTNASC'] = pd.to_datetime(sinasc_raw['DTNASC'])

# Extraímos o mês da coluna DTNASC
sinasc_raw['Mes'] = sinasc_raw['DTNASC'].dt.month

# Filtramos para o mês de março
marco_data = sinasc_raw[sinasc_raw['Mes'] == 3]

# Agrupamos por município e aplicamos as agregações
estatisticas_municipio = (
    marco_data.groupby('munResNome')
    .agg({
        'DTNASC': 'count',  # Contar o número de ocorrências de DTNASC (número de nascimentos)
        'QTDFILVIVO': ['mean', 'max', 'min'],  # Estatísticas do número de filhos vivos
        'IDADEPAI': ['mean', 'max', 'min']   # Estatísticas das idades dos pais
    })
)

# Encontramos o município com o maior número de nascimentos em março
municipio_mais_nascimentos = estatisticas_municipio['DTNASC'].idxmax()
num_nascimentos = estatisticas_municipio['DTNASC'].max()

# Filtramos as estatísticas para o município com mais nascimentos
estatisticas_municipio_max = estatisticas_municipio.loc[municipio_mais_nascimentos]

# Exibimos os resultados
print(f"Município com mais nascimentos em março: {municipio_mais_nascimentos}")
print(f"Número de nascimentos em março: {num_nascimentos}")
print("Estatísticas de QTDFILVIVO e IDADEPAI para o município com mais nascimentos:")
print(estatisticas_municipio_max)
```
