```python
import pandas as pd
import requests

# 1) seu código aqui
sinasc = pd.read_csv('SINASC_RO_2019.csv')
print(sinasc.shape)
sinasc.drop_duplicates().shape
# Não há duplicados
```

```python
# 2) seu código aqui

#contando dados faltantes
sinasc.isna().sum()
```

```python
# 3) seu código aqui
colunas_desejadas= ['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO', 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']
sinasc_filtro = sinasc[colunas_desejadas]
sinasc_filtro.head(10)
```

```python
# 4) seu código aqui
sinasc_filtro.dropna(subset=['APGAR5'], inplace=True)
```

```python
sinasc_filtro.head(5)
```

```python
sinasc_filtro.shape #contando o número de linhas
```

```python
sinasc_filtro.isna().sum() # contando o número de missing
```

```python
# 5) seu código aqui
import numpy as np

# Substituir valores 9 por NaN nas variáveis específicas
sinasc_filtro[['ESTCIVMAE', 'CONSULTAS']] = sinasc_filtro[['ESTCIVMAE', 'CONSULTAS']].replace(9, np.nan)
sinasc_filtro[['ESTCIVMAE', 'CONSULTAS']].isna().sum()
```

```python
# 6) Seu código aqui
# É necessário o entedimento da função se eu quisesse substituir por 1000 faria o código assim: sinasc_filtro['QTDFILVIVO'].fillna(1000, inplace=True)
# Substituir valores missing em QTDFILVIVO por 0

sinasc_filtro['QTDFILVIVO'].fillna(0, inplace=True)
```

```python
sinasc_filtro['QTDFILVIVO'].value_counts()
```

```python
# 7) seu código aqui
# identificando quais colunas referente a questão estão com missing
sinasc_filtro.isna().sum()
```

```python
# Aplicando o método
sinasc_filtro['GRAVIDEZ'].fillna(method='ffill', inplace=True)
sinasc_filtro['GESTACAO'].fillna(method='ffill', inplace=True)
```

```python
sinasc_filtro.isna().sum()
```

```python
# Preferi preencher, pois percebo que a variação não é grande entre os dados e necessariamente, no caso em questão, os dados devam existir.
```

```python
# 8) seu código aqui

#Criando coluna e bool para dados da mesma
sinasc_filtro.loc[sinasc_filtro['APGAR5'] >= 8, 'Respiração' ] = 'Normal'
sinasc_filtro.loc[(sinasc_filtro['APGAR5'] <= 7) & (sinasc_filtro['APGAR5'] >= 6), 'Respiração' ] = 'Asfixia leve'
sinasc_filtro.loc[(sinasc_filtro['APGAR5'] <= 5) & (sinasc_filtro['APGAR5'] >=4), 'Respiração'] = 'Asfixia moderada'
sinasc_filtro.loc[(sinasc_filtro['APGAR5'] <= 0) & (sinasc_filtro['APGAR5'] >=3), 'Respiração'] = 'asfixia severa'
```
