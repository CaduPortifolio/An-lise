#### 2) Vamos montar um metadados

1. Crie um dataframe com os nomes de cada variável e o tipo de dados de cada variável.
2. Adicione uma coluna nesse *dataframe* chamada "qtd_categorias" e coloque nela o número de categorias correspondente de cada variável.
    Dica:
        1. inicie uma lista vazia
        2. faça um for ao longo dos nomes das variáveis,
        3. conte o número de categorias dessa variável
        4. acumule essa informação de 3. na lista que você criou em 1.
        5. No final, essa lista pode ser atribuída à nossa variável.

````python
# Inicializando uma lista para armazenar os metadados
metadados = []

# Iterando sobre as colunas do DataFrame
for coluna in df.columns:
    tipo_dado = df[coluna].dtype
    if tipo_dado == 'object':
        # Para variáveis categóricas, contamos o número de categorias únicas
        qtd_categorias = df[coluna].nunique()
    else:
        # Para variáveis não categóricas, o número de categorias é 0
        qtd_categorias = 0

    # Adiciona as informações à lista de metadados
    metadados.append({
        'nome_variavel': coluna,
        'tipo_dado': tipo_dado,
        'qtd_categorias': qtd_categorias
    })

# Criando o DataFrame de metadados
df_metadados = pd.DataFrame(metadados)

print(df_metadados)
````

````python
df_metadados
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
# Identificando as variáveis qualitativas (categóricas) que são do tipo `object`
variaveis_qualitativas = df.select_dtypes(include=['object']).columns

# Criando variáveis dummy para as variáveis qualitativas
df_dummies = pd.get_dummies(df, columns=variaveis_qualitativas)

# Exibindo o DataFrame final com as variáveis dummy
df_dummies.head(5)
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

Criando a tabela cruzada.

````python
tc = pd.crosstab(index=[df['possui_email'], df['posse_de_veiculo']], columns=df['mau'])
````

Calculando a proporção de 'mau' = 1 para cada combinação

````python
tc_proporcoes = tc.div(tc.sum(axis=1), axis= 0)
````

Exibindo o resultado da tabela cruzada com proporções.

````python
tc_proporcoes
````
