####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
#1) Seu código aqui
url_sp = 'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=SP&tipo_interessado=Estado'
````

````python
def consultar_dados(uf, tipo_interessado):
    url = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado={tipo_interessado}'
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()  # Assume que a resposta é JSON
        df = pd.DataFrame(data['items'])  # Ajuste conforme a estrutura da resposta
        return df
    else:
        print(f'Erro na consulta: {response.status_code}')
        return None
````

````python
#testando função
url_sp = consultar_dados('SP', 'Estado')
url_sp
````

````python
# 2) Seu código aqui
contar_solicitacoes_arquivadas = lambda df: len(df[df['status'] == 'Arquivado por decurso de prazo']) if df is not None else 0

# Exemplo de uso:
df_mg = consultar_dados('MG', 'Estado')
quantidade_arquivada = contar_solicitacoes_arquivadas(df_mg)
print(f'Quantidade de solicitações arquivadas por decurso de prazo em MG: {quantidade_arquivada}')
````

````python
# 3) Seu código aqui
df_bahia_municipio = consultar_dados('BA', 'Munic%C3%ADpio')
solicitacoes_deferidas = df_bahia_municipio[df_bahia_municipio['status'] == 'Deferido']
municipio_com_mais_solicitacoes = solicitacoes_deferidas['interessado'].value_counts()
municipio_com_mais_solicitacoes
````

````python
# 4) Seu código aqui
def salvar_dados_bahia():
    df = consultar_dados('BA', 'Estado')
    if df is not None:
        # Salvando o DataFrame em um arquivo CSV
        df.to_csv('solicitacoes_bahia_estado.csv', index=False)
        print("Dados salvos em 'solicitacoes_bahia_estado.csv'")
    else:
        print("Não foi possível salvar os dados.")


salvar_dados_bahia()
````
