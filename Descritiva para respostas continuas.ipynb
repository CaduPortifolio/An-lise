````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
````

1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
df = pd.read_csv('previsao_de_renda.csv')
df.head(5)
````

````python
ax = sns.pointplot(x='posse_de_imovel', y='renda',
                   data=df, dodge=True, errorbar=('ci', 90))
plt.xlabel('Tem Imóvel')
plt.ylabel('Renda')
plt.title('Renda de pessoas que têm ou não imóveis')

plt.show()
````

````python
ax_veiculo = sns.pointplot(x='posse_de_veiculo', y='renda',
                           data=df, dodge=True, errorbar=('ci', 95))

plt.xlabel('Tem veículo')
plt.ylabel('Renda')
plt.title('Renda de pessoas que possuem ou não veículos')

plt.show()
````

````python
ax_im = sns.pointplot(x='qtd_filhos', y='renda', hue='posse_de_imovel',
              data=df, dodge=True, errorbar=('ci', 90), palette={True:'green', False:'red'})

plt.xlabel('Quantidade de Filhos')
plt.ylabel('Renda')
plt.title('Renda de pessoas com filhos e imóveis')

plt.show()
````

````python
ax_v = sns.pointplot(x="qtd_filhos", y="renda", hue="posse_de_veiculo",
                     data=df, dodge=True, errorbar=("ci", 90), palette={True: "green", False: "red"})

plt.xlabel("Quantidade de Filhos")
plt.ylabel("Renda")
plt.title("Renda de Pessoas Com Filhos e Veículos")

plt.show()
````

````python
ax_e = sns.pointplot(
    x="educacao",  # Alterado para 'educacao'
    y="renda",
    hue="posse_de_veiculo",
    data=df,
    dodge=True,
    errorbar=("ci", 90),
    palette={True: "green", False: "red"},
)

plt.xlabel("Escolaridade")  # Alterado para 'Escolaridade'
plt.ylabel("Renda")
plt.title("Renda de Pessoas Com Escolaridade e Veículos")
# Rotacionar os rótulos do eixo x em 90 graus
plt.xticks(rotation=90)
plt.show()
````

````python


ax_e = sns.pointplot(
    x="educacao",
    y="renda",
    hue="posse_de_imovel",
    data=df,
    dodge=True,
    errorbar=("ci", 90),
    palette={True: "green", False: "red"},
)

plt.xlabel("Escolaridade")
plt.ylabel("Renda")
plt.title("Renda de Pessoas Com Escolaridade e Imóveis")
# Rotacionar os rótulos do eixo x em 90 graus
plt.xticks(rotation=90)

plt.show()
````
