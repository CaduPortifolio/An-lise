#### 1)  crie uma série do pandas a partir de uma lista com os dados abaixo:

Em um estudo sobre alteração na temperatura global, A NASA disponibiliza dados de diferenças de temperatura média da superfície terrestre relativos às médias de temperatura entre 1951 e 1980. Os dados originais podem ser vistos no site da NASA/GISS, e estão dispostos a cada década na tabela abaixo.

|ano|anomalia termica|
|:-:|:----:|

| 1920 | -0.27 |
| 1940 | 0.12 |
| 1960 | -0.03 |
| 1980 | 0.26 |
| 2000 | 0.40 |
| 2020 | 1.02 |

Crie uma séries do Pandas a partir de uma lista com esses dados.

#### 5) Nomeie os índices das linhas com inteiros de 1 a 20, e as colunas com os nomes "x1", "x2", e "x3" respectivamente.

```python
#seu código aqui
# Gerar o ndarray com números pseudo-aleatórios
arr = np.random.normal(100, 10, (20, 3))

# Transformar o ndarray em um DataFrame
df = pd.DataFrame(arr, columns=['x1', 'x2', 'x3'])

# Nomear os índices das linhas com inteiros de 1 a 20
df.index = range(1, 21)

# Exibir o DataFrame
df
```

#### 6) No DataFrame do exercício 5, crie uma nova coluna como sendo a média das três colunas, e dê a ela o nome de "media" (não recomendo colocar acentos em nomes de variáveis).

```python
# seu código aqui
# Calcular a média das três colunas e adicionar como uma nova coluna 'media'
df['media'] = df.mean(axis=1)
df
```

#### 7) No DataFrame do exercício 6, crie uma nova coluna chamada "log_med", contendo o logaritmo natural da média calculada no exercício 6 <br>

```python
# seu código aqui
# Calcular o logaritmo natural da média e adicionar como uma nova coluna 'log_med'
df['log_med'] = np.log(df['media'])
df
```
